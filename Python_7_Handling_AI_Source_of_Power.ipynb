## Write Text File

Pada program sebelumnya, kita hanya bisa me-read text file. Untuk me-write  file text kita perlu menambahkan argument **mode** pada function `open()`. Berikut adalah mode yang bisa kita gunakan untuk me-write file:

<br>

Mode | Keterangan
-|-
`r+` | read dan meniban konten file paling depan dengan data baru. File yang dibuka harus ada dalam directory terlebih dahulu.
`w` | mode untuk **meniban** seluruh konten file dengan data baru. Jika file yang dibuka tidak ada, maka file akan dibuat.
`a` | mode untuk menambah konten file dengan data baru. Jika file yang dibuka tidak ada, maka file akan dibuat.
`w+` | sama seperti mode `w` tetapi juga bisa melakukan read
`a+` | sama seperti mode `a` tetapi juga melakukan read

<br>

Jika argument mode tidak di-passing, function `open()` akan secara otomatis memilih mode `r` (read-only).

Coba jalankan program berikut kemudian analisis hasilnya!

Hubungannya ialah binary merupakan bentuk lain dari pixel. Seperti yang kita tahu, komputer hanya bisa memproses binary, sementara pixel bisa bernilai dari 0 sampai dengan 255 (bukan binary). Supaya bisa memproses pixel, komputer perlu mengkonversi pixel menjadi binary terlebih dahulu.

## Tugas 3

1. Cari image apapun di internet.
2. Read dan tampilkan image.
3. Cari tahu ukuran tensor-nya.
4. Lakukan operasi aritmatika apapun pada image.
5. Tampilkan image hasil operasi aritmatika.
6. Simpan image hasil operasi aritmatika dalam folder 'latihan' dengan nama 'latihan3.png'

```python
import requests
from PIL import Image
from io import BytesIO
import numpy as np
import matplotlib.pyplot as plt
import os

# 1. Cari dan ambil gambar dari internet
url = 'https://pingintau.id/wp-content/uploads/2023/06/kucing.jpg'  # Ganti URL ini dengan URL gambar
response = requests.get(url)
img = Image.open(BytesIO(response.content))

# 2. Tampilkan gambar
plt.imshow(img)
plt.title("Original Image")
plt.axis('off')
plt.show()

# 3. Cari ukuran tensor dari gambar
img_array = np.array(img)
print(f"Shape of the image tensor: {img_array.shape}")

# 4. Lakukan operasi aritmatika pada gambar (contoh: mengalikan dengan skalar)
img_modified_array = img_array * 0.5  # Operasi mengurangi brightness
img_modified_array = img_modified_array.astype(np.uint8)  # Pastikan kembali dalam tipe data uint8

# 5. Tampilkan gambar hasil operasi aritmatika
plt.imshow(img_modified_array)
plt.title("Modified Image (Brightness Reduced)")
plt.axis('off')
plt.show()

# 6. Simpan gambar hasil operasi aritmatika dalam folder 'latihan' dengan nama 'latihan3.png'
os.makedirs('latihan', exist_ok=True)
img_modified = Image.fromarray(img_modified_array)
img_modified.save('latihan/latihan3.png')
print("Image saved as 'latihan/latihan3.png'")
```

## Tugas 4

1. Download file dari link berikut https://ai.stanford.edu/~amaas/data/sentiment/aclImdb_v1.tar.gz kemudian simpan dalam folder 'latihan' dengan nama 'latihan4.tar.gz'

2. Extract file tersebut kemudan simpan hasil ekstrak di folder 'latihan'.

```python
import os
import requests
import tarfile

# URL file yang akan didownload
url = "https://ai.stanford.edu/~amaas/data/sentiment/aclImdb_v1.tar.gz"
# Folder tempat menyimpan file
folder = 'latihan'
# Nama file setelah didownload
file_name = 'latihan4.tar.gz'
# Path lengkap file
file_path = os.path.join(folder, file_name)

# Buat folder 'latihan' jika belum ada
os.makedirs(folder, exist_ok=True)

response = requests.get(url, stream=True)
with open(file_path, 'wb') as file:
    for chunk in response.iter_content(chunk_size=1024):
        if chunk:
            file.write(chunk)
print(f"File downloaded and saved as {file_path}")

# Extract file
print(f"Extracting file {file_path}...")
with tarfile.open(file_path, 'r:gz') as tar:
    tar.extractall(path=folder)
print(f"File extracted to {folder}")
```
